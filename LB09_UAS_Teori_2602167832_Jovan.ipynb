```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#rhombus code

def rhombus_equation(x,a):
    return a - np.abs(x)

#membuat dataset
x = np.linspace(-10.0, 10.0, 200)
a_value1 = 10
y = rhombus_equation(x,a_value1)
y = np.hstack([y,-y])
x = np.hstack([x,-x])

x1 = np.linspace(-9.0, 9.0, 200)
a_value2 = 50
y1 = rhombus_equation(5*x1,a_value2)
y1 = np.hstack([y1,-y1])
x1 = np.hstack([x1,-x1])

df1 = pd.DataFrame(np.vstack([y,x]).T,columns=['x1','x2'])
df1['Y'] = 0
df2 = pd.DataFrame(np.vstack([y1,x1]).T,columns=['x1','x2'])
df2['Y'] = 1

# Menggabungkan DataFrames menggunakan pd.concat
df = pd.concat([df1, df2],ignore_index=True)
df.head()
```

```python
# run in a different cell
plt.scatter(y,x)
plt.scatter(y1,x1)
```

```python
df.isnull().sum()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
input_df=df.drop('Y',axis=1)
output_df=df['Y']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_df, output_df, test_size = 0.2, random_state = 0)
```

```python
#Linear Kernel model 1
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score

svc = make_pipeline(SVC(kernel='linear',random_state=0))
svc.fit(x_train,y_train)
y_predict = svc.predict(x_test)

print(f"Classification Report for Linear kernel Model 1\n")
print(classification_report(y_true=y_test,y_pred=y_predict,target_names=['0','1']))

acccuracyLinear1 = round(accuracy_score(y_true=y_test,y_pred=y_predict),2)
recallLinear1_class0 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=0),2)
recallLinear1_class1 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=1),2)
precisionLinear1_class0 = round(precision_score(y_test,y_predict,pos_label=0),2)
precisionLinear1_class1 = round(precision_score(y_test,y_predict,pos_label=1),2)
f1Linear1_class0 = round(f1_score(y_test,y_predict,pos_label=0),2)
f1Linear1_class1 = round(f1_score(y_test,y_predict,pos_label=1),2)
```

```python
#Linear Kernel model 2

svc = SVC(kernel='linear',gamma=0.001,C=100, random_state=1)
svc.fit(x_train,y_train)
y_predict = svc.predict(x_test)

print(f"Classification Report for Linear kernel Model 2\n")
print(classification_report(y_true=y_test,y_pred=y_predict,target_names=['0','1']))

acccuracyLinear2 = round(accuracy_score(y_true=y_test,y_pred=y_predict),2)
recallLinear2_class0 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=0),2)
recallLinear2_class1 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=1),2)
precisionLinear2_class0 = round(precision_score(y_test,y_predict,pos_label=0),2)
precisionLinear2_class1 = round(precision_score(y_test,y_predict,pos_label=1),2)
f1Linear2_class0 = round(f1_score(y_test,y_predict,pos_label=0),2)
f1Linear2_class1 = round(f1_score(y_test,y_predict,pos_label=1),2)
```

```python
#Radial Basis Function (RBF) Kernel Model 1:
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.pipeline import make_pipeline

svc = make_pipeline(SVC(kernel='rbf',random_state= 1))
svc.fit(x_train,y_train)
y_predict = svc.predict(x_test)

print(f"Classification Report for Radial Basis Function (RBF) Kernel Model 1:\n")
print(classification_report(y_true=y_test,y_pred=y_predict,target_names=['0','1']))

acccuracyRBF1 = round(accuracy_score(y_true=y_test,y_pred=y_predict),2)
recallRBF1_class0 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=0),2)
recallRBF1_class1 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=1),2)
precisionRBF1_class0 = round(precision_score(y_test,y_predict,pos_label=0),2)
precisionRBF1_class1 = round(precision_score(y_test,y_predict,pos_label=1),2)
f1RBF1_class0 = round(f1_score(y_test,y_predict,pos_label=0),2)
f1RBF1_class1 = round(f1_score(y_test,y_predict,pos_label=1),2)
```

```python
#Radial Basis Function (RBF) Kernel Model 2:
from sklearn.svm import SVC
from sklearn.metrics import classification_report
from sklearn.pipeline import make_pipeline

svc = make_pipeline(SVC(kernel='rbf',random_state= 1,gamma=0.001,C=100))
svc.fit(x_train,y_train)
y_predict = svc.predict(x_test)

print(f"Classification Report for Radial Basis Function (RBF) Kernel Model 2:\n")
print(classification_report(y_true=y_test,y_pred=y_predict,target_names=['0','1']))

acccuracyRBF2 = round(accuracy_score(y_true=y_test,y_pred=y_predict),2)
recallRBF2_class0 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=0),2)
recallRBF2_class1 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=1),2)
precisionRBF2_class0 = round(precision_score(y_test,y_predict,pos_label=0),2)
precisionRBF2_class1 = round(precision_score(y_test,y_predict,pos_label=1),2)
f1RBF2_class0 = round(f1_score(y_test,y_predict,pos_label=0),2)
f1RBF2_class1 = round(f1_score(y_test,y_predict,pos_label=1),2)
```

```python
df = pd.read_csv("HeartFailurePredictionDataset.csv")
```

```python
df.info()
```

```python
df
```

```python
#check NA
df.isnull().sum()
```

**Featuring Engineering**

```python
binary_encode={"Sex": {"M":0,"F":1},"ExerciseAngina": {"N":0,"Y":1}}
label_encode={"ChestPainType": {"TA":0,"ATA":1,"NAP":2,"ASY":3},
              "RestingECG": {"Normal":0,"ST":1,"LVH":2},
              "ST_Slope":{"Down":0,"Flat":1,"Up":2}}
```

```python
df = df.replace(binary_encode)
df = df.replace(label_encode)
```

```python
df
```

```python
#split
input_df = df.drop(['HeartDisease'],axis=1)
output_df = df['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(input_df, output_df, test_size = 0.2, random_state = 0)
```

```python
#DecisionTree
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score

dt = DecisionTreeClassifier(random_state=0)
dt.fit(x_train,y_train)
y_predict = dt.predict(x_test)

print("classification report for  DecisionTree\n")
print(classification_report(y_true=y_test,y_pred=y_predict,target_names=['0','1']))

acccuracy1 = round(accuracy_score(y_true=y_test,y_pred=y_predict),2)
recall1_class0 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=0),2)
recall1_class1 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=1),2)
precision1_class0 = round(precision_score(y_test,y_predict,pos_label=0),2)
precision1_class1 = round(precision_score(y_test,y_predict,pos_label=1),2)
f11_class0 = round(f1_score(y_test,y_predict,pos_label=0),2)
f11_class1 = round(f1_score(y_test,y_predict,pos_label=1),2)
```

```python
#Bagging with 50 models
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,recall_score,f1_score,precision_score

bagging = BaggingClassifier(n_estimators=100,oob_score=True)
bagging.fit(x_train,y_train)
y_predict = bagging.predict(x_test)

print("classification report for bagging with 100 models using DecisionTree\n")
print(classification_report(y_true=y_test,y_pred=y_predict,target_names=['0','1']))

acccuracy2 = round(accuracy_score(y_true=y_test,y_pred=y_predict),2)
recall2_class0 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=0),2)
recall2_class1 = round(recall_score(y_true=y_test,y_pred=y_predict,pos_label=1),2)
precision2_class0 = round(precision_score(y_test,y_predict,pos_label=0),2)
precision2_class1 = round(precision_score(y_test,y_predict,pos_label=1),2)
f12_class0 = round(f1_score(y_test,y_predict,pos_label=0),2)
f12_class1 = round(f1_score(y_test,y_predict,pos_label=1),2)
```

```python
Model = ['DecisionTree','Bagging w/ DT using 100 estimators']

Accuracy = [acccuracy1,acccuracy2]
Support_Accuracy = [184,184]

Precision_0 = [precision1_class0,precision2_class0]
Precision_1 = [precision1_class1,precision2_class1]

Recall_0 = [recall1_class0,recall2_class0]
Recall_1 = [recall1_class1,recall2_class1]

f1_0 = [f11_class0,f12_class0]
f1_1 = [f11_class1,f12_class1]

support0 = [77,77,77,77]
support1 = [107,107,107,107]
```

```python
classification_df = pd.DataFrame({"Model":Model,"Accuracy":Accuracy,"Support":Support_Accuracy,"Recall class 0":Recall_0,"Precision class 0":Precision_0,"F1 score class 0":f1_0,"Recall class 1":Recall_1,
                                  "Precision class 1":Precision_1,"F1 score class 1":f1_1}).T

classification_df
```

**based on the models , we can conclude that the bagging with 100 models of DecisionTree performs superior than the normal DecisionTree model in every aspect**
